```python
# :SMART LIFE SIMULATOR & AI COMPANION:

# STEP: 1
#=============================
# Making a Class named [User]
#=============================

class User:
    def __init__(self, name):
        self.name = name
        self.energy = 100
        self.day = 1
        self.tasks = []
        self.completed_tasks = []

# STEP: 2
#=======================================
# Create a function named Task Manager
#=======================================

def TaskManager_add_task(user):
    name = input("Enter task name: ")
    priority = int(input("Enter priority (1-5): "))
    user.tasks.append({"name": name, "priority": priority, "done": 0})
    print("Task added successfully.")

def TaskManager_show_tasks(user):
    if not user.tasks:
        print("No tasks available.")
    else:
        print("Today's Tasks:")
        for i, t in enumerate(user.tasks, 1):
            print(i, t["name"], "| Priority:", t["priority"])

def TaskManager_complete_task(user):
    name = input("Enter task name to complete: ")
    for t in user.tasks:
        if t["name"] == name:
            user.completed_tasks.append(t)
            user.tasks.remove(t)
            t["done"] += 1
            user.energy -= t["priority"] * 5
            print("Task completed.")
            return
    print("Task not found.")

# STEP: 3
#=======================================
# Create a function named AI companion
#=======================================

def AICompanion_predict_next_task(user):
    if not user.tasks:
        print("AI: No task to suggest.")
        return
    best = user.tasks[0]
    for t in user.tasks:
        if t["priority"] < best["priority"]:
            best = t
    print("AI Suggests Next Task:", best["name"])
    print("AI Advices you:\n1) Follow Up Your Passion.\n2) Always Carefull About Your Work.\n3) Always Think Positive.")

# STEP: 4
#===========================================
# Create a function named Resource Manager
#===========================================

def ResourceManager_allocate_energy(user):
    if not user.tasks:
        print("No tasks to allocate energy.")
        return
    per_task = user.energy // len(user.tasks)
    print("Energy Allocation:")
    for t in user.tasks:
        print(t["name"], "→", per_task)

# STEP: 5
#===============================================
# Create a function named Random Event Manager
#===============================================

def RandomEventManager_daily_event(user):
    events = [
        ("Unexpected free time", +3),
        ("Extra workload", -3),
        ("Motivational boost", +5),
        ("Low energy day", -5)
    ]
    index = user.day % len(events)
    event = events[index]
    user.energy += event[1]
    print("Daily Event:", event[0], "| Energy:", event[1])

# STEP: 6
#======================================
# Create a function named Day Summary
#======================================

def Main_end_day_summary(user):
    print("\n--- End of Day", user.day, "---")
    print("Completed tasks:", len(user.completed_tasks))
    print("Remaining tasks:", len(user.tasks))
    print("Energy left:", user.energy)
    user.completed_tasks.clear()
    user.day += 1

# STEP: 7
#======================
# Runs the Main loops
#======================
user = User(input("Enter your name: "))
print("\nWelcome to Smart Life Simulator & AI Companion,", user.name)

while True:
    print("\nDay", user.day, "| Energy:", user.energy)
    RandomEventManager_daily_event(user)

    print("\n1 Add Task")
    print("2 Show Tasks")
    print("3 AI Suggestions")
    print("4 Complete Task")
    print("5 Allocate Energy")
    print("6 End Day")
    print("7 Exit\n")

    choice = input("Choose an option: ")

    if choice == "1":
        TaskManager_add_task(user)
    elif choice == "2":
        TaskManager_show_tasks(user)
    elif choice == "3":
        AICompanion_predict_next_task(user)
    elif choice == "4":
        TaskManager_complete_task(user)
    elif choice == "5":
        ResourceManager_allocate_energy(user)
    elif choice == "6":
        Main_end_day_summary(user)
    elif choice == "7":
        print("Exiting system. Goodbye!")
        break
    else:
        print("Invalid option.")
```
